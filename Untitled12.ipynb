```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("predict.csv")
df.head()
```

```python
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.duplicated().sum()
```

```python
df['CampaignChannel'].value_counts()
```

```python
ab_data = df[df['CampaignChannel'].isin(['Email','Social'])]
```

```python
conversion_rate = ab_data.groupby('CampaignChannel')['Conversion'].mean()
print(conversion_rate)
```

```python
sns.barplot(x='CampaignChannel', y='Conversion', data=ab_data)

plt.title("Conversion Rate by Campaign Channel")
plt.show()
```

```python
from statsmodels.stats.proportion import proportions_ztest

email = ab_data[ab_data['CampaignChannel']=='Email']['Conversion']
social = ab_data[ab_data['CampaignChannel']=='Social']['Conversion']

success = [email.sum(), social.sum()]
total = [email.count(), social.count()]

z_stat, p_value = proportions_ztest(success, total)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='CampaignType', y='Conversion', data=ab_data)

plt.title("Conversion Rate by Campaign Type")
plt.show()
```

```python
from statsmodels.stats.proportion import proportions_ztest

groupA = ab_data[ab_data['CampaignType']=='Conversion']['Conversion']
groupB = ab_data[ab_data['CampaignType']=='Awareness']['Conversion']

success = [groupA.sum(), groupB.sum()]
total = [groupA.count(), groupB.count()]

z_stat, p_value = proportions_ztest(success, total)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```
